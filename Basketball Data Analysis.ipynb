```python
def myplot(playerlist):
    col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Orange","CarmeloAnthony":"Yellow","DwightHoward":"Green","ChrisBosh":"Blue","ChrisPaul":"Magenta","KevinDurant":"Brown","DerrickRose":"Purple","DwayneWade":"Crimson"}
    Mk = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"d","CarmeloAnthony":"D","DwightHoward":"1","ChrisBosh":"<","ChrisPaul":"2","KevinDurant":"^","DerrickRose":"3","DwayneWade":"X"}
    plt.figure(figsize=(15, 6))
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c=col[name], ls="--", marker=Mk[name], ms='9', label=name)
    # Adding a Legend
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    # Add dates replacing range
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.show()
```

```python
# Add matrix to look at represented by "data"

def myplot(data,playerlist):
    col = {"KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Orange","CarmeloAnthony":"Yellow","DwightHoward":"Green","ChrisBosh":"Blue","ChrisPaul":"Magenta","KevinDurant":"Brown","DerrickRose":"Purple","DwayneWade":"Crimson"}
    Mk = {"KobeBryant":"s","JoeJohnson":"o","LeBronJames":"d","CarmeloAnthony":"D","DwightHoward":"1","ChrisBosh":"<","ChrisPaul":"2","KevinDurant":"^","DerrickRose":"3","DwayneWade":"X"}
    plt.figure(figsize=(15, 8))
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=col[name], ls="--", marker=Mk[name], ms='9', label=name)
    # Adding a Legend
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    # Add dates replacing range
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.show()
```
